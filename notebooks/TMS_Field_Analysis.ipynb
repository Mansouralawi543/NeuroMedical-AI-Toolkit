# TMS Field Analysis and Visualization
This notebook demonstrates the simulation of induced E-fields in neural tissue during Transcranial Magnetic Stimulation (TMS).

```python
import numpy as np
import matplotlib.pyplot as plt
from src.modeling.tms_simulation import TMSSimulator
```

```python
sim = TMSSimulator()
coil_pos = np.array([0, 0, 0.05])  # 5cm above origin

# Create target grid for visualization
x = np.linspace(-0.05, 0.05, 20)
y = np.linspace(-0.05, 0.05, 20)
X, Y = np.meshgrid(x, y)
Z = np.zeros_like(X) # 5cm below coil

grid = np.column_stack([X.flatten(), Y.flatten(), Z.flatten()])
e_fields = sim.calculate_e_field(coil_pos, 1e8, grid)

# Reshape for quiver plot
U = e_fields[:, 0].reshape(X.shape)
V = e_fields[:, 1].reshape(X.shape)
M = np.sqrt(U**2 + V**2)
```

```python
plt.figure(figsize=(10, 8))
plt.quiver(X, Y, U, V, M, cmap='viridis')
plt.title('Induced E-field Vectors in Tissue Plane')
plt.colorbar(label='Field Magnitude (V/m)')
plt.xlabel('X (m)')
plt.ylabel('Y (m)')
plt.show()
```
